```python
import mujoco

# Replace with the actual main scene or robot XML file name inside the repo
mjcf_path = "../arms/openarm_mujoco/v2/cell.xml"

model = mujoco.MjModel.from_xml_path(mjcf_path)
data = mujoco.MjData(model)

print(f"Total Generalized Coordinates (nq): {model.nq}")
print(f"Total Degrees of Freedom (nv): {model.nv}")

print("\nMJCF Joint Ordering:")
for i in range(model.njnt):
    joint_name = mujoco.mj_id2name(model, mujoco.mjtObj.mjOBJ_JOINT, i)
    # Get the address/index of this joint's position in the qpos array
    qpos_adr = model.jnt_qposadr[i]
    print(f"Joint ID {i} -> Name: '{joint_name}' (Maps to data.qpos[{qpos_adr}])")
```
